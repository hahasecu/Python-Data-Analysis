### analyze the age column

```python
# function age_int to convert string age into integers
def age_int(age_str):
    if pd.isnull(age_str):
        return None
    else:
        age = age_str.split(' ')[0]
        age = age.replace('+','')
        
    return int(age)
```

```python
data_yes['int_age'] = data_yes['Age'].apply(age_int)
```

```python
data_yes['int_age'].describe()
```

```python
data_yes['How much total combined money did all members of your HOUSEHOLD earn last year?'].value_counts()
```

### analyze household income

```python
# function to covert income string into numerical data
def int_income(income_str):
    if pd.isnull(income_str):
        return None
    income_str = income_str.split(' ')[0]
    if income_str == 'Prefer':
        return None
    else:
        income_str = income_str.replace('$', '').replace(',', '')
    return int(income_str)
```

```python
data_yes['int_income'] = data_yes['How much total combined money did all members of your HOUSEHOLD earn last year?'].apply(int_income)
data_yes['int_income'].describe()
```

### explore correlation between int_income and travel distance
### It's safe to hypothesize that people earning less money could be younger, and would travel to their parent's houses for Thanksgiving. People earning more are more likely to have Thanksgiving at their house as a result.

```python
data_yes['How far will you travel for Thanksgiving?'].value_counts()
```

```python
income_150000 = data_yes[data_yes['int_income']>150000]
```

```python
income_150000['How far will you travel for Thanksgiving?'].value_counts()
```

## Foundings
### The possibility of spend Thanksgiving in their homes is higher among those who earn more than 150000.
